```python
import numpy as np

class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.connectivity = len(value)
        
class  Graph:
    def __init__(self,dic):  #the data type of arguement is dictionary
        self.Nodes = []
        for key in dic.keys():
            self.Nodes.append(Node(key,dic[key]))
    
    def ave_connect(self):
        ave_list = []
        for n in self.Nodes:
            ave_list.append(n.connectivity)
        ave_list = np.array(ave_list)
        print("The value of average connectivity is %s."%(round(np.mean(ave_list),1))) 
    
    def max_connect(self):
        max_list = []
        for n in self.Nodes:
            max_list.append(n.connectivity)
        max_list = np.array(max_list)
        max_value = np.max(max_list)
        key_list = []
        for n in self.Nodes:
            if n.connectivity == max_value:
                key_list.append(n.key)
        print("The value of maximum connectivity is %s. The list of node objects having this level of connectivity is %s."%(max_value,key_list))
    
    def min_connect(self):
        min_list = []
        for n in self.Nodes:
            min_list.append(n.connectivity)
        min_list = np.array(min_list)
        min_list = np.unique(min_list)
        if np.min(min_list) == 0:
            min_value = np.sort(min_list)[1]
        else:
            min_value = np.min(min_list)
            
        key_list = []
        for n in self.Nodes:
            if n.connectivity == min_value:
                key_list.append(n.key)
        print("The value of minimum connectivity is %s. The list of node objects having this level of connectivity is %s."%(min_value,key_list))
    
        
    def disconnect(self):
        dis_list = []
        for n in self.Nodes:
            if n.connectivity == 0:
                dis_list.append(n.key)
        if len(dis_list) == 0:
            print("There are no islands.")
        else:
            print("The disconnected nodes are: %s."%dis_list)
            
    def summary(self):
        self.ave_connect()
        self.max_connect()
        self.min_connect()
        self.disconnect()

def gal ():
    #read the file
    url = input("Please enter the file path: ")
    f = open(url,"r")
    
    listings = []
    dic = {}
    for x in f:
        listings.append(x)
        #print(listings)
    for n in range(1,len(listings),2):
        key = listings[n].split(' ')[0]
        value = listings[n+1].split(' ')
        value[-1] = value[-1][:-1]
        dic[key] = value
    f.close()
    return dic

#TEST
gal1 = gal()
G1 = Graph(gal1)
'''
G1.ave_connect()
G1.max_connect()
G1.min_connect()
G1.disconnect()
'''
G1.summary()
```
