```python
#根据老师的教案，因为一个圆比一个正方形的ratio为pi/4，所以pi=4*|C|/|S
# 在一个正方形范围内，随机无数个点既在圆内又在正方形内的占比，就是ratio=|C|/|S|。
```

```python
#首先加载package
import random #注意这个不是numpy的random，是random这个packge的random
import math
import matplotlib.pyplot as plt
```

```python
# 先计算原本pi是多少
pi = math.pi
pi
```

```python
# 开始编码需要生成多少次随机点才能得到置信率为0.00001的𝜋值

n = 0 #掉落在圆里面的点的个数
d = 0 #掉落在正方形内的点的个数
simulating = True #？这个作为哨兵？不太懂，我理解是作为一个开启的开关
while simulating: #不用加 = True
    x = random.random() #用random package里面的random method生成0-1的value
    y = random.random() #同样生成一个y值，那么这x和y值作为一个点的坐标，代表了一个随机点生成了
    if x**2 + y**2 <= 1.0: #生成点以后，如果这个点在圆内或圆上
        n += 1 #圆内点的个数+1
    d += 1 #d无论如何都+1，因为这个点一定在正方形内
    EstimatePi = 4 * n * 1.0 /d #计算4*ratio
    print(EstimatePi)
    if abs(EstimatePi - pi) / pi <= 0.00001:#abs()是算绝对值，整个公式算p-value
        print("Draws needed: " + d)
        break #当p-value小于0.00001，得到我们想要的值的时候，停止这个循环迭代
```

```python
n = 0
d = 0
EstimatedPiList = [] #定义一个List
xs = []
ys = []
simulating = True # use as a sentinel
while simulating:
    x = random.random()
    y = random.random()
    xs.append(x)
    ys.append(y)
    if x**2 + y**2 <= 1.0:
        n += 1
    d += 1
    EstimatedPi = 4 * n * 1./d
    
    print(EstimatedPi)
    EstimatedPiList.append(EstimatedPi) #定义一个list不断加上新生成的EstimatedPi
    if abs(EstimatedPi-pi) / pi <= 0.00001:
        print("Draws needed: ", d)
        break
```
