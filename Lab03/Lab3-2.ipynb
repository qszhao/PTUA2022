```python
#Method 1
import random
import math
import matplotlib.pyplot as plt
pi = math.pi

def simulate_pie():
    cc = float(input("Please input a convergence criterion:"))
    num = int(input("Please input a sentinel value that limits the number of draws:"))
    n = 0 #number of points falling in the unit circle
    d = 0 #number of points falling in the unit square
    ratios = []
    simulating = 0 #use as a sentinel
    while simulating < num:
        x = random.random()
        y = random.random()
        if x**2+y**2<=1.0:
            n += 1
        d += 1
        ratio = 4*n*1./d
        print(ratio)
        ratios.append(ratio)
        if abs(ratio-pi) / pi <=cc:
            print("Draws needed:",d)
            break
        simulating += 1
        if simulating == num:
            print("Draw limit was exceeded.")
            ratio_s = 'percent:{:.2%}'.format((ratio-pi) / pi)
            print("The percentage difference between the current estimate of Pi is %s. The value of Pi is %s."%(ratio_s,pi))
            plt.plot(ratios)
simulate_pie()   
```

```python
#Method 2
import random
import math
import numpy as np
pi = math.pi

def simulate_pie_10(cc):
    number = []
    for i in range(10):
        n = 0 #number of points falling in the unit circle
        d = 0 #number of points falling in the unit square
        simulating = True #use as a sentinel
        while simulating:
            x = random.random()
            y = random.random()
            if x**2+y**2<=1.0:
                n += 1
            d += 1
            ratio = 4*n*1./d
            #print(ratio)
            if abs(ratio-pi) / pi <=cc:
                #print("Draws needed:",d)
                number.append(d)
                i += 1
                break
    print(number)
    avg = round(sum(number)/len(number),2)
    std = round(np.std(number, ddof = 1),2)
    print("The average number of draws required from 10 runs is: %s.(%s)"%(avg,cc))
    print("The standard deviation of the number of draws required from 10 runs is: %s.(%s)"%(std,cc))
        
simulate_pie_10(0.01) 
simulate_pie_10(0.001)
simulate_pie_10(0.0001)
simulate_pie_10(0.00001)
```
