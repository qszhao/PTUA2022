```python
#3 is_palindrome
def is_palindrome(word):
    length = len(word)
    loop = int(length/2)
    flag = 0
    for i in range(1,loop+1):
        if word[::i] == word[::-i]:
            continue
        else:
            print("This word is not palindrome.")
            flag = 1
            break
    if flag == 0:
        print("This word is palindrome.")
        
is_palindrome('aabaac')    
is_palindrome('aabaa')
```

```python
#lab03-2
#1
import random
import math
pi = math.pi
criterion = input("Please enter the convergence criterion（a positive number）:")
sentinel = input("Please enter a sentinel value(a integer):")
criterion = float(criterion)
sentinel = int(sentinel)

n = 0 # number of points falling in the unit circle
d = 0 # number of points falling in the unit square
simulating = True # use as a sentinel
times = 0
while simulating:
    times += 1
    x = random.random()
    y = random.random()
    if x**2 + y**2 <= 1.0:
        n += 1
    d += 1
    ratio = 4 * n * 1./d
    print(ratio)
    if abs(ratio-pi) / pi <= criterion:
        print("Draws needed: ", d)
        break
    if times > sentinel:
        difference = abs(ratio-pi) / pi
        percentage =  "%.2f%%" % (difference * 100)
        print("Draw limit was exceeded, the percentage difference is ",percentage)
        break
```
