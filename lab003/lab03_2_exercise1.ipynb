```python
import random
import math
import matplotlib.pyplot as plt
pi = math.pi
n = 0
d = 0
simulating = True
while simulating:
    x = random.random()
    y = random.random()
    if x**2 + y**2 <= 1.0:
        n += 1
    d += 1
    ratio = 4 * n * 1./d
    print(ratio)
    if abs(ratio-pi) / pi <= 0.00001:
        print("Draws needed: ", d)
        break
```
